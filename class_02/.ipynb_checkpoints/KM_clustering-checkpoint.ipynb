```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
mean_01 = np.array([1.0, 3.0])
mean_02 = np.array([-1.0, 0.0])

cov_01 = np.array([[0.8, 0.3], [0.3, 0.9]])
cov_02 = np.array([[0.9, 0.4], [0.4, 0.8]])

dist_01 = np.random.multivariate_normal(mean_01, cov_01, 400)
dist_02 = np.random.multivariate_normal(mean_02, cov_02, 400)

print dist_01.shape, dist_02.shape
```

```python
plt.figure(0)
plt.xlim(-6, 6)
plt.ylim(-8, 8)
plt.scatter(dist_01[:, 0], dist_01[:, 1], color='red')
plt.scatter(dist_02[:, 0], dist_02[:, 1], color='green')

plt.show()
```

```python
data = np.zeros((dist_01.shape[0] + dist_02.shape[0], dist_01.shape[1]))
data[:dist_01.shape[0]] = dist_01
data[dist_01.shape[0]:] = dist_02

print data.shape, data.mean(axis=0)

data_std = (data - data.mean(axis=0))/data.std(axis=0)
print data_std.mean(axis=0), data_std.std(axis=0)
```

```python
plt.figure(0)
plt.xlim(-6, 6)
plt.ylim(-8, 8)
plt.scatter(data_std[:, 0], data_std[:, 1])
plt.show()
```

```python
def distance(x1, x2):
    return np.sqrt(((x1-x2)**2).sum())
```

```python
centers = {}
k = 3

for ix in range(k):
    d = {
        'coords': np.random.random((2,)),
        'pts': []
    }
    if ix == 2:
        d['coords'] = np.array([6, -6])
    centers['c' + str(ix)] = d

print centers
```

```python
colors = ['red', 'green', 'blue', 'yellow', 'cyan']
```

```python
N = data_std.shape[0]

for ix in range(N):
    dist = []
    for ic in range(k):
        d = distance(data_std[ix], centers['c' + str(ic)]['coords'])
        dist.append(d)
    dist = np.asarray(dist)
    c = dist.argmin()
    centers['c' + str(c)]['pts'].append(data_std[ix])

plt.figure(0)
for ik in range(k):
    c = np.asarray(centers['c' + str(ik)]['pts'])
    cc = centers['c' + str(ik)]['coords']
    print cc
    try:
        plt.scatter(c[:, 0], c[:, 1], color=colors[ik])
    except:
        pass
    plt.scatter(cc[0], cc[1], color='black')
    
plt.figure(1)    

for ik in range(k):
    c = np.asarray(centers['c' + str(ik)]['pts'])
    if not len(c) is 0:
        centers['c' + str(ik)]['coords'] = c.mean(axis=0)
    cc = centers['c' + str(ik)]['coords']
    print cc
    try:
        plt.scatter(c[:, 0], c[:, 1], color=colors[ik])
    except:
        pass
    plt.scatter(cc[0], cc[1], color='black')
    centers['c' + str(ik)]['pts'] = []
plt.show()
```
