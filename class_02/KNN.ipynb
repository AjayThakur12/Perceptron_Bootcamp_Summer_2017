```python
mean_01 = np.array([1, 1])
cov_01 = np.array([[2.9, 1.4], [1.7, 1.5]])

mean_02 = np.array([4, 7])
cov_02 = np.array([[2, -0.7], [-0.7, 1.5]])

data_01 = np.random.multivariate_normal(mean_01, cov_01, 500)
data_02 = np.random.multivariate_normal(mean_02, cov_02, 500)
print data_01.shape
print data_02.shape
```

```python
plt.figure(0)

plt.scatter(data_01[:, 0], data_01[:, 1])
plt.scatter(data_02[:, 0], data_02[:, 1], color='red')

plt.show()
```

```python
all_data = np.zeros((1000, 3))

all_data[:500, :2] = data_01

all_data[500:, :2] = data_02
all_data[500:, -1] = 1

np.random.shuffle(all_data)
print all_data[:10]
```

```python
split = int(0.8 * all_data.shape[0])

train = all_data[:split]
test = all_data[split:]

print train.shape, test.shape
```

```python
print all_data.mean(axis=0)
```

```python
def knn(data, testing, k=5):
    N = data.shape[0]
    dist = []
    
    for ix in range(N):
        d = distance(data[ix, :2], testing)
        dist.append([d, data[ix, -1]])
    
    neighbours = sorted(dist, key=lambda x: x[0])
    k_neighbours = neighbours[:k]
    
    dist_and_labels = np.array(k_neighbours)
    labels = dist_and_labels[:, -1]
    
    freq = np.unique(labels, return_counts=True)
    return freq[0][freq[1].argmax()]


def distance(x1, x2):
    return np.sqrt(((x1-x2)**2).sum())

# print distance(np.array([1, 1]), np.array([0, 0]))
```

```python
z = np.array([5, 4])

out = knn(train, z, k=9)

print out
```

```python
record = []

for ix in range(test.shape[0]):
    record.append(knn(train, test[ix, :2], k=5))
record = np.asarray(record)

print record
```
