```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import cv2
import keras
from keras.datasets import cifar10
```

```python
ds = cifar10.load_data()
```

```python
im = ds[0][0][100, :, :, :]
im.shape
```

```python
plt.imshow(im)
```

```python
K = np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]])
print K
```

```python
res = np.zeros((im.shape[0] - K.shape[0] + 1, im.shape[1] - K.shape[1] + 1))
print res.shape
```

```python
img = cv2.cvtColor(im, cv2.COLOR_RGB2GRAY)
print img.shape
```

```python
plt.imshow(img, cmap='gray')
```

```python
for ix in range(res.shape[0]):
    for iy in range(res.shape[1]):
        im_patch = img[ix:ix+K.shape[0], iy:iy+K.shape[1]]
        # print np.sum(K*im_patch)
        res[ix, iy] = np.sum(im_patch * K)
```

```python
plt.imshow(res, cmap='gray')
```
