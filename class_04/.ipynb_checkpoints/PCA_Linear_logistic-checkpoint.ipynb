```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import sklearn
from sklearn.decomposition import PCA
```

```python
ds = pd.read_csv('../train.csv')
all_data = ds.values[:5000, :]
print all_data.shape
```

```python
X = all_data[:, 1:]
y = all_data[:, 0]
```

```python
pca = PCA(n_components=X.shape[1])

pca.fit(X)
X_ = pca.transform(X)
```

```python
plt.figure(0)
plt.imshow(np.cov(X_.T), cmap='gray', interpolation='none')

plt.figure(1)
plt.imshow(np.cov(X.T), cmap='gray', interpolation='none')
plt.show()
```

```python
c = np.cov(X_.T)
print c.min(), c.max()
# print c[100:110, 100:110]
# print '-'*80
c1 = np.cov(X.T)
# print c1[100:110, 100:110]
```

```python
plt.figure(0)
plt.plot(pca.explained_variance_ratio_)

plt.show()
```

```python
f = np.cumsum(pca.explained_variance_ratio_)
print f.shape
```

```python
plt.figure(0)
plt.ylim(0, 1.01)
plt.plot(f)
plt.show()
```

```python
nd = 2
print "Information Gain in " + str(nd) + " dimensions", f[nd]*100

for ix in range(f.shape[0]):
    if f[ix] >= 0.98:
        print ix, f[ix]
        break
```

```python
def f(x):
    w = np.array([-1.0, 2.3])
    return w[0]*x[0] + w[1]*x[1] + 4.2 + 2*(2*np.random.random() - 1)
```

```python
N = 1000
X_data = 20*np.random.random((N, 2))
y_data = np.zeros((N,))

for ix in range(N):
    y_data[ix] = f(X_data[ix])
```

```python
print lr_01.score(X_data[800:], y_data[800:])
print lr_01.coef_
print lr_01.intercept_
```

# Logistic Regression

$$
h(x) = \sigma (y(x))
$$

$$
y(x) = \sum w_i x_i
$$

```python
mean_01 = np.array([1, 1])
cov_01 = np.array([[2.9, 1.4], [1.7, 1.5]])

mean_02 = np.array([4, 7])
cov_02 = np.array([[2, -0.7], [-0.7, 1.5]])

data_01 = np.random.multivariate_normal(mean_01, cov_01, 500)
data_02 = np.random.multivariate_normal(mean_02, cov_02, 500)
print data_01.shape
print data_02.shape
```

```python
plt.figure(0)

plt.scatter(data_01[:, 0], data_01[:, 1])
plt.scatter(data_02[:, 0], data_02[:, 1], color='red')

plt.show()
```

```python
all_data = np.zeros((1000, 3))

all_data[:500, :2] = data_01

all_data[500:, :2] = data_02
all_data[500:, -1] = 1

np.random.shuffle(all_data)
print all_data[:10]
```

```python
split = int(0.8 * all_data.shape[0])

train = all_data[:split]
test = all_data[split:]

print train.shape, test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr_02 = LogisticRegression(verbose=0)
lr_02.fit(train[:, :-1], train[:, -1])
```

```python
lr_02.score(test[:, :-1], test[:, -1])
```

```python
print lr_02.coef_
print lr_02.intercept_
```

# Logistic Regression in MNIST with PCA

```python
ds = pd.read_csv('../train.csv')
all_data = ds.values[:5000, :]
print all_data.shape
```

```python
import datetime
```

```python
X = all_data[:, 1:]
y = all_data[:, 0]
print y
```

```python
Y = np.asarray((y%2 == 0), dtype='uint8')
print Y
```

```python
start = datetime.datetime.now()
lr01 = LogisticRegression()
lr01.fit(X[:4500], Y[:4500])
print "Time Taken:", datetime.datetime.now() - start
```

```python
lr01.score(X[4500:], Y[4500:])
```

```python
# Apply PCA
pca = PCA(n_components=148)
XX = pca.fit_transform(X)
```

```python
start = datetime.datetime.now()
lr01 = LogisticRegression()
lr01.fit(XX[:4500], Y[:4500])
print "Time Taken:", datetime.datetime.now() - start
```

```python
lr01.score(XX[4500:], Y[4500:])
```

```python
def check_pca_accuracy(X_data, y_data, n_components):
    pca = PCA(n_components=n_components)
    XX_data = pca.fit_transform(X_data)
    lr = LogisticRegression(n_jobs=-1)
    lr.fit(XX_data[:4500], y_data[:4500])
    return lr.score(XX_data[4500:], y_data[4500:])
```

```python
dims = [2, 25, 32, 64, 128, 256, 343, 512, 784]
acc = []
for dx in dims:
    start = datetime.datetime.now()
    acc.append(check_pca_accuracy(X, y, dx))
    print "Time Taken for dimension {0}:".format(dx), datetime.datetime.now() - start
```

```python
plt.figure(0)
plt.plot(acc)
```
